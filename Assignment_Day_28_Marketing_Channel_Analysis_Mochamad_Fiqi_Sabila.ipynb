# 4. Segmentasi Pelanggan dengan RFM Analysis

## A. Copy df

```python
df_superstore = df.copy()
df_superstore.head()
```

```python
df_superstore.info()
```

## B. RFM Table

```python
ref_date = df_superstore['Order Date'].max()
```

```python
rfm = df_superstore.groupby('Customer ID').agg({
    'Order Date': lambda x: (ref_date - x.max()).days,  # Recency
    'Order ID': pd.Series.nunique,                     # Frequency
    'Sales': 'sum'                                      # Monetary
}).reset_index()
```

```python
rfm.columns = ['Customer ID', 'Recency', 'Frequency', 'Monetary']
```

```python
rfm.head()
```

## C. R Score, F Score, M Score

```python
# 1. Skor Recency: semakin kecil recency, skor makin tinggi
rfm['R_Score'] = pd.qcut(rfm['Recency'], q=3, labels=[3, 2, 1])

# 2. Skor Frequency: semakin sering, skor makin tinggi
rfm['F_Score'] = pd.qcut(rfm['Frequency'].rank(method='dense'), q=3, labels=[1, 2, 3])

# 3. Skor Monetary: semakin besar sales, skor makin tinggi
rfm['M_Score'] = pd.qcut(rfm['Monetary'].rank(method='dense'), q=3, labels=[1, 2, 3])

# 4. Gabungkan skor menjadi RFM_Score
rfm['RFM_Score'] = rfm['R_Score'].astype(str) + rfm['F_Score'].astype(str) + rfm['M_Score'].astype(str)

# 5. Lihat hasilnya
rfm.head()
```

```python
plt.figure(figsize=(14,6))

# Hitung distribusi RFM_Score
score_counts = rfm['RFM_Score'].value_counts().sort_index()  # sort_index agar urutan numerik

# Plot barplot
sns.barplot(
    x=score_counts.index,
    y=score_counts.values
)

# Tambahkan label di atas bar
for i, val in enumerate(score_counts.values):
    plt.text(i, val + 2, str(val), ha='center', va='bottom', fontsize=9)

# Format visual
plt.title("Distribusi Kombinasi RFM Score")
plt.xlabel("RFM Score (RFM)")
plt.ylabel("Jumlah Customer")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

## D. Mapping Segmen

```python
def map_rfm_segment(score):
    if score in ['331', '332', '333']:
        return 'Champion'
    elif score in ['321', '322', '323']:
        return 'Loyal Customers'
    elif score in ['311', '312', '313']:
        return 'Potential Loyalists'
    elif score in ['231', '232', '233']:
        return 'Promising'
    elif score in ['221', '222', '223']:
        return 'New Customers'
    elif score in ['211', '212', '213']:
        return 'Need Attention'
    elif score in ['131', '132', '133']:
        return 'At Risk'
    elif score in ['121', '122', '123']:
        return 'About to Sleep'
    elif score in ['111', '112', '113']:
        return 'Lost'
    else:
        return 'Others'
```

```python
#mapping ke dataframe
rfm['RFM_Segment_Label'] = rfm['RFM_Score'].apply(map_rfm_segment)
```

```python
#distribusinya
rfm['RFM_Segment_Label'].value_counts()
```

```python
plt.figure(figsize=(12,6))

# Hitung distribusi dan atur urutan dari terbesar ke terkecil
segment_counts = rfm['RFM_Segment_Label'].value_counts().sort_values(ascending=False)

# Barplot
sns.barplot(
    x=segment_counts.index,
    y=segment_counts.values
)

# Tambahkan label angka di atas batang
for i, val in enumerate(segment_counts.values):
    plt.text(i, val + 1, str(val), ha='center', va='bottom')

# Format visual
plt.title("Distribusi RFM Segmentasi Pelanggan")
plt.xlabel("Segment Label")
plt.ylabel("Jumlah Customer")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```
